Load the datasets

```python
from datasets import load_dataset, DatasetDict, load_dataset_builder

eval_data = DatasetDict()

eval_data = load_dataset("mozilla-foundation/common_voice_11_0", "sv-SE")
```

```python
eval_data = eval_data.remove_columns(["accent", "age", "client_id", "down_votes", "gender", "locale", "segment", "up_votes"])
```

```python
eval_data = eval_data["test"]
```

```python
eval_data = eval_data.select(range(100))
```

### Generate y_pred

Get the pipeline

```python
from transformers import pipeline
pipe = pipeline(model="fimster/whisper-small-sv-SE-NST")
```

```python
def update_path(path):
  return path[0:119] + 'sv-SE_test_0//' + path[119:]
```

```python
y_pred = []
i = 1

for path in eval_data["path"]:
  print(f"Proccessing {i} of {len(eval_data)}!")
  updated_path = update_path(path)
  y_pred.append(pipe(updated_path)["text"])
  i += 1
```

```python
y = eval_data["sentence"]
```

Nomralize

```python
import re

y_pred_norm = [re.sub(r'[^\w]', ' ', s).lower() for s in y_pred]
```

```python
y_norm = [re.sub(r'[^\w]', ' ', s).lower() for s in y]
```

```python
y_pred_norm
```

Compute WER

```python
from evaluate import load
wer = load("wer")
wer_score = wer.compute(predictions=y_pred_norm, references=y_norm)
```

```python
print("NST WER= ", wer_score)
```

Mozilla WER

```python
from transformers import pipeline
pipe_mozilla = pipeline(model="fimster/whisper-small-sv-SE")
```

```python
def update_path(path):
  return path[0:119] + 'sv-SE_test_0/' + path[119:]
```

```python
eval_data[0]["path"]
```

```python
y_pred_mozilla = []
i = 1

for path in eval_data["path"]:
  print(f"Proccessing {i} of {len(eval_data)}!")
  updated_path = update_path(path)
  print("New PATH= ", updated_path)
  y_pred_mozilla.append(pipe_mozilla(updated_path)["text"])
  i += 1
```

```python
y_mozilla = eval_data["sentence"]
```

```python
import re

y_pred_mozilla_norm = [re.sub(r'[^\w]', ' ', s).lower() for s in y_pred_mozilla]
```

```python
y_mozilla_norm = [re.sub(r'[^\w]', ' ', s).lower() for s in y_mozilla]
```

```python
from evaluate import load
wer = load("wer")
wer_score_mozilla = wer.compute(predictions=y_pred_mozilla_norm, references=y_mozilla_norm)
```

```python
print("MOZILLA WER= ", wer_score_mozilla)
```
